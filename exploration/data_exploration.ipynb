# Data exploration for comprehension

This notebook is used to explore the provided data, which is made of:

- `data/kc_house_data.csv` – Data for training the model.  Each row corresponds to a sold home.  `price` is the target column, the rest can be used as features if appropriate.
- `data/zipcode_demographics.csv` – Additional demographic data from the U.S. Census which are used as features.  This data should be joined to the primary home sales using the `zipcode` column.
- `data/future_unseen_examples.csv` – This file contains examples of homes to be sold in the future.  It includes all attributes from the original home sales file, but not the `price`, `date`, or `id`. It also does not include the demographic data.

# kc_house_data

The training dataset. Where `price` is the target

```python
import pandas as pd
import os 
import sys
from pathlib import Path

# print options for pandas
pd.set_option('display.max_columns', None)

# Add project root to path - notebook is in exploration/, go up one level
project_root = Path.cwd().parent
sys.path.insert(0, str(project_root))

def LoadDataframe(file_directory: str,file_name: str) -> pd.DataFrame:
    file_path = os.path.join(project_root, file_directory, file_name)
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        return df
    else:
        raise FileNotFoundError(f"The file {file_path} does not exist.")
```

```python
file_folder = "data"
file_name_house_data = "kc_house_data.csv"

house_data_df = LoadDataframe(file_folder, file_name_house_data)
print(house_data_df.shape)
house_data_df.head(10)
```

# Exploring the relations between tables

```python
# Check if all zipcodes in house_data_df are in demographics_df
house_zips = set(house_data_df['zipcode'].unique())
demo_zips = set(demographics_df['zipcode'].unique())
unseen_zips = set(unseen_df['zipcode'].unique())

print(f"Total unique zipcodes in house_data_df: {len(house_zips)}")
print(f"Total unique zipcodes in demographics_df: {len(demo_zips)}")
print(f"Total unique zipcodes in unseen_df: {len(unseen_zips)}")

missing_zips = house_zips - demo_zips

print(f"Missing zipcodes (in house data but not in demographics): {len(missing_zips)}")

missing_unseen_zips = unseen_zips - demo_zips

print(f"Missing zipcodes in unseen_df (in unseen data but not in demographics): {len(missing_unseen_zips)}")
```
